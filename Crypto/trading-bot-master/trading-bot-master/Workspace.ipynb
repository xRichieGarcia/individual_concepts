```python
x_range = np.arange(len(chartData))
closeData = np.zeros(len(chartData), dtype=float)

maxClose = 0;
for i in x_range:
    closeData[i] = chartData[i]['close']
    if (closeData[i] > maxClose):
        maxClose = closeData[i]

plt.scatter(x_range, closeData, alpha=0.5)
plt.ylabel("close price")
plt.title("Histogram of closing price of " + pair  + " over past year")
x1,x2,y1,y2 = plt.axis()
ymax = 1.2 * maxClose
plt.axis((x1, x2, 0, ymax))

plt.show()
```
